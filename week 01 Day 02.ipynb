## **Comments**

Comments in Python are lines in the code that are not executed by the computer. They are written to help explain what the code does, making it easier for others (or your future self) to understand. Comments are a crucial part of writing readable, maintainable code.

*   **Python supports single-line comments using the # symbol**
*   **Multi-line comments using triple quotes (''' or """).**
*   

```python

'''
-----Welcome to my program-------
Here you will see how I did addition of two numbers
 boths numbers are integer
'''

a=6
b=5
c=a+b

print(f"Sum of a and b numbers are:{c}") #The sum of two numbers will be printed



```

## **Variables**
Variables are like labeled containers used to store data values in a program. In Python, you can create a variable by simply assigning a value to a name using the = sign, like x = 5. Once created, a variable can be used and updated throughout the code. Variables make it easier to store, reuse, and manage data while writing programs.

```python
# Variables and Assignment

age = 25
name = "John"
print("The age of a person is:", age)
print("The name of the person is:", name)
print("The name of the person is:"+name) # string can be concatination,'+'
```

```python
# Display values
print("Name:", name)
print("Age:", age)
```

```python
# Changing variable values
#age = age + 1

age=30

age+=1

print("Age next year:", age)
```

## **Expression**

An expression is any combination of variables, values, and operators that Python can evaluate (or "compute") to produce another value. For example, 2 + 3 is an expression that evaluates to 5. Expressions can involve arithmetic operations (+, -, *, /), string concatenation, and even logical comparisons. They show how logic and calculations are performed in programs.

```python
# Arithmetic Expressions
a = 10
b = 3
c=a+b
d=a-b
e=a*b
print("Addition:", c)
print("Subtraction:", d)
print("Multiplication:", e)
```

```python
print("Division:", a / b) # ceiling ....
print("Integer Division:", a // b) # floor
print("Modulus:", a % b)
print("Exponent:", a ** b) #  check power
```

```python
# String expressions
first = "Hello"
second = "World"
print(first + second)
```

```python
First_name="Ahmad"
Last_name="Ali"
print(First_name+" "+Last_name+" !")

```

## **Data Types**

Data types represent the kinds of values that variables can hold in Python. The most common data types include:

integers (int),

*   floating-point numbers (float)
*   strings (str)
*   booleans (bool)
*   lists
*   tuples
*   dictionaries
*   sets.
Understanding data types is essential because they determine what operations can be performed on a piece of data. For example, you can add two integers, but you cannot directly add a string and an integer without converting one of them.

```python
# Integer

a = "42"
print(type(a), a)
```

```python
# Float
pi = 3.1415
print(type(pi), pi)
```

```python
# String
text = "Python"
print(type(text))
```

```python
# Boolean

is_active = True
print(type(is_active), is_active)

```

```python
# List
fruits = ["apple", "banana", "cherry"]
print(type(fruits), fruits)

```

```python
# Dictionary
person = {"name": "Alice", "age": 30}
print(type(person), person)

```

```python
# Tuple
coordinates = (10, 20)
print(type(coordinates), coordinates)

```

```python
# Set
unique_numbers = {1, 2, 3, 3, 2}
print(type(unique_numbers), unique_numbers)
```
